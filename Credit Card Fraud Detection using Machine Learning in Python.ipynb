```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')


from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
#loading dataset to pandas as DataFrame 

credit_card_data = pd.read_csv('archive (1)/creditcard.csv')
```

```python
# This dataset is highly unbalanced because more than 99% of data in the one particular class, so we cannot feed this data into out ml model
# If we do this then ml model cannot able to recognize the fraudulent data because we have very less data points for this case so whenever you gave new datapoint it will predict that it is nornal transaction.
# 0 -normal transactions
# 1 -fraudulent transactions
```

```python
#seperating the data for analysis i.e ( normal and fraud transactions)

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statisticals measures of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both of this transactions for mean of all cols in legit and fraud transactions

credit_card_data.groupby('Class').mean()
```

```python
# Under-Sampling
# Build a sample dataset containing similar distribution of normal transactions and Fraudulent transactions
# so we have number of fraudulent transactions = 492 
# thus we will take the 492 rows from legit transaction ramdomly
# so dataset will become uniform 

legit_sample = legit.sample(n = 492)  #ramdom sampling of dataset 
```

```python
# Concatenating 2 dataframes

new_dataset = pd.concat([legit_sample,fraud], axis =0)  #axis 0 is to add this values row wise and fraud row will come after the legit_sample row.
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into features and targets

X = new_dataset.drop(columns = 'Class', axis =1) #remove class column
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into training and testing data

X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y ,random_state = 2)  

# test_size = 0.2 , represents 20% data for testing
# stratify=Y: Keeps the same class ratio (e.g. fraud vs. non-fraud) in both train and test sets.
# random_state=2: Ensures the split is the same every time you run the code (for reproducibility)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training

# Logistic Regression

model = LogisticRegression()
```

```python
# Training the logistic regression model with training data

model.fit(X_train, Y_train)
```

```python
# Model Evaluation based of Accuracy score

# accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data :', training_data_accuracy)
```

```python
# accuracy on testing data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data :', test_data_accuracy)
```

```python
# Assuming y_pred is your prediction for X_test
y_pred = model.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(Y_test, y_pred))

print("\nClassification Report:")
print(classification_report(Y_test, y_pred))
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)
rf_preds = rf_model.predict(X_test)

print("Random Forest:")
print(confusion_matrix(Y_test, rf_preds))
print(classification_report(Y_test, rf_preds))
```
